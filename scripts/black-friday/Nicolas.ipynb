```python
# Notes of Nicolas

import pandas

## From .csv > data frame
blackFridayDataFrame = pandas.read_csv('../data/black-friday/BlackFriday.csv')

blackFridayDataFrame
```

```python
## Extract all 'Product_ID' from the data frame
collectionProductId = blackFridayDataFrame['Product_ID']

## Select only the unique 'Product_ID'
collectionDistinctProductId = collectionProductId.drop_duplicates()

print('Counting all product-id: '+str(collectionProductId.count()))
print('Counting all unique product-id: '+str(collectionDistinctProductId.count()))
print('First unique product-id: '+str(collectionDistinctProductId.sort_values(ascending=False)[0]))
print('Last unique product-id: '+str(collectionDistinctProductId.sort_values(ascending=False)[collectionDistinctProductId.count()-1]))

print()

## Extract all 'City_Category'
collectionCityCategory = blackFridayDataFrame['City_Category']
collectionDistinctCityCategory = collectionCityCategory.drop_duplicates()

print('All city categories:')
print(collectionDistinctCityCategory.sort_values())

print()

## Extract all 'Occupation'
collectionOccupation = blackFridayDataFrame['Occupation']
collectionDistinctOccupation = collectionOccupation.drop_duplicates()

print('All occupation values:')
print(collectionDistinctOccupation.sort_values())

print()

## Extract all 'Stay_In_Current_City_Years'
collectionCurrentStayInCity = blackFridayDataFrame['Stay_In_Current_City_Years']
collectionDistinctCurrentStayInCity = collectionCurrentStayInCity.drop_duplicates()

print('All Stay_In_Current_City_Years values:')
print(collectionDistinctCurrentStayInCity.sort_values())

print()

## Extract all ''
collectionAgeGroups = blackFridayDataFrame['Age']
collectionDistinctAgeGroups = collectionAgeGroups.drop_duplicates()

print('All Age-groups:')
print(collectionDistinctAgeGroups.sort_values())
```

```python
## Check what's really in a product_category, filter on first Product_ID 'P00069042'

blackFridayDataFrame[blackFridayDataFrame.Product_ID == collectionDistinctProductId.sort_values(ascending=False)[0]]
```
